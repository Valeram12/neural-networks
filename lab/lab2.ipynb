```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/Valeram12/neural-networks/main/lab1/Housing.csv')

df= df.head(100)
df = df.dropna()

print("Датасет:\n", df)
```

```python
# 1. Гістограма цін
plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Гістограма цін')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Кругова діаграма кількості спалень
plt.figure(figsize=(8, 6))
bedroom_counts = df['bedrooms'].value_counts()
plt.pie(bedroom_counts, labels=bedroom_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightblue', 'lightyellow'])
plt.title('Кругова діаграма кількості спалень')
plt.show()
```

```python
# 3. Стовпчаста діаграма: Кількість спалень проти ціни
plt.figure(figsize=(8, 6))
plt.bar(df['bedrooms'], df['price'], color='lightgreen', edgecolor='black')
plt.title('Стовпчаста діаграма: Кількість спалень проти ціни')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# 4. Точкова діаграма: Площа проти ціни
plt.figure(figsize=(8, 6))
plt.scatter(df['area'], df['price'], color='lightblue', edgecolor='black')
plt.title('Точкова діаграма: Площа проти ціни')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# 5. Лінійний графік - Лінійний графік цін за індексом
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['price'], color='orange', marker='o', linestyle='-', markersize=5)
plt.title('Лінійний графік цін за індексом')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

```python
# 6. Розподіл цін на нерухомість
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True, color='blue')
plt.title('Розподіл цін на нерухомість')
plt.xlabel('Ціна')
plt.ylabel('Частота')
plt.show()
```

```python
# 7. Розмір площі vs Ціна
plt.figure(figsize=(8, 6))
sns.regplot(x='area', y='price', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Регресійна лінія: Залежність ціни від розміру площі')
plt.xlabel('Площа (кв.м)')
plt.ylabel('Ціна')
plt.show()
```

```python
# 8. Кількість кімнат vs Ціна
plt.figure(figsize=(8, 6))
sns.stripplot(x='bedrooms', y='price', data=df, jitter=True, hue='furnishingstatus', palette='Set2')
plt.title('Залежність ціни від кількості спалень')
plt.xlabel('Кількість спалень')
plt.ylabel('Ціна')
plt.show()
```

```python
# 9. Кількість місць для паркування vs Ціна
plt.figure(figsize=(8, 6))
sns.barplot(x='parking', y='price', data=df)
plt.title('Залежність ціни від кількості паркувальних місць')
plt.xlabel('Паркувальні місця')
plt.ylabel('Ціна')
plt.show()
```

```python
# 10
pivot_table = df.pivot_table(values='price', index='bedrooms', columns='furnishingstatus', aggfunc='mean')

plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='Blues')
plt.title('Pivot Table: Середня ціна залежно від кількості спалень та меблювання')
plt.xlabel('Меблювання')
plt.ylabel('Кількість спалень')
plt.show()
```
