```python
import pandas as pd
import numpy as np

def insert_null(df):
    # Вставка рядків з NaN через кожні 5 рядків
    rows_to_insert = []
    for i in range(5, len(df) + len(df) // 5, 6):
        nan_row = pd.Series([np.nan] * len(df.columns), index=df.columns)
        rows_to_insert.append((i, nan_row))

    for idx, nan_row in rows_to_insert:
        df = pd.concat([df.iloc[:idx], pd.DataFrame([nan_row]), df.iloc[idx:]]).reset_index(drop=True)
    return df


df = pd.read_csv('https://raw.githubusercontent.com/Valeram12/neural-networks/main/lab1/Housing.csv')
print("Датасет:\n", df)

df_wrong = insert_null(df)
print("\nДатасет з відсутніми даними:\n", df_wrong)

missing_data = df_wrong.isna().sum()
print("\nВідсутні дані у стовпцях:\n", missing_data)

df_filled = df_wrong.fillna(0)
print("\nЗамінені відсутні дані на 0:\n", df_filled.head(10))

df_dropped = df_wrong.dropna()
print("\nВидалені рядки з NaN:\n", df_dropped.head(10))

df['total_rooms'] = df['bedrooms'] + df['bathrooms']
print("\nДатасет з доданим стовпцем 'total_rooms':\n", df.head(10))

aircon_count = df['airconditioning'].value_counts().reset_index()
print("\nКількість будинків з кондиціонерами та без:\n", aircon_count)

furnishing_status_count = df['furnishingstatus'].value_counts().reset_index()
print("\nКількість будинків за статусом меблів:\n", furnishing_status_count)
```
