```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Valeram12/neural-networks/refs/heads/main/lab3/howpop_train.csv")
df.T
#task 3 Виведіть дані так як показано на зображенні.
```

```python
#task 4 Видаліть данні, що завершуються на _lognom
df = df.filter(regex='^(?!.*_lognorm$)')
df
```

```python
#task 5 Застосуйте для виводу статистичної інформації функцію DataFrame.describe()
df.describe()
```

```python
#task 6 Стовпець published (час публікації) містить рядки. Щоб ми могли працювати з цими даними як з датою/ часом публікації, перетворіть їх до типу datetime:
df['published'] = pd.to_datetime([i for i in df.published])
df.info()
df.head(5)
```

```python
#task 7 Створіть кілька стовпців на основі даних про час публікації:
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]
df['dayofweek'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
df
```

#task 1

В якому році та місяці було найбільше публікацій?

- Березень 2016

- Березень 2015

- Квітень 2015

- Квітень 2016

```python
#task1
import plotly.express as px
df['year_month'] = df['published'].dt.to_period('M').astype(str)
publication_counts = df['year_month'].value_counts().sort_index()
max_index = publication_counts.idxmax()
fig = px.bar(
    x=publication_counts.index,
    y=publication_counts.values,
    labels={'x': 'Рік та місяць', 'y': 'Кількість публікацій'},
    title='Кількість публікацій по місяцях',
    text=publication_counts.values,
    color=['red' if x == max_index else 'normal' for x in publication_counts.index]
)

fig.update_xaxes(tickangle=45)
fig.show()

most_published_month = publication_counts.idxmax()
print(f'Місяць з найбільшою кількістю публікацій: {most_published_month}')
```

# Task 2

Проаналізуйте публікації в місяці з попереднього питання. Оберіть один або кілька варіантів.

- Один або кілька днів сильно виділяються із загальної картини

- На Хабрі (habrahabr.ru) завжди більше статей, ніж на Гіктаймсі (geektimes.ru)

- По суботах на Гіктаймсі та Хабрі публікують приблизно однакове число статей

```python
#task2
data_month = df[df['published'].dt.to_period('M').astype(str) == most_published_month]
data_month['day'] = data_month['published'].dt.day
data_month['weekday'] = data_month['published'].dt.day_name()

daily_counts = data_month['day'].value_counts().sort_index()

weekday_counts = data_month['weekday'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    fill_value=0
)

import plotly.express as px

fig_daily = px.bar(
    x=daily_counts.index,
    y=daily_counts.values,
    labels={'x': 'День місяця', 'y': 'Кількість публікацій'},
    title=f'Кількість публікацій по дням у {most_published_month}',
    text=daily_counts.values
)

fig_daily.update_traces(texttemplate='%{text:.0f}', textposition='outside')
fig_daily.show()
```

```python
fig_weekday = px.bar(
    x=weekday_counts.index,
    y=weekday_counts.values,
    labels={'x': 'День тижня', 'y': 'Кількість публікацій'},
    title=f'Кількість публікацій по дням тижня у {most_published_month}',
    text=weekday_counts.values
)

fig_weekday.update_traces(texttemplate='%{text:.0f}', textposition='outside')
fig_weekday.show()
```

По цьому графіку можна побачити, що сама більша кількість публікацій була зроблена по вівторках

---



```python
platform_counts = data_month['domain'].value_counts()

fig_platform = px.bar(
    x=platform_counts.index,
    y=platform_counts.values,
    labels={'x': 'Платформа', 'y': 'Кількість публікацій'},
    title=f'Кількість публікацій на платформах у {most_published_month}',
    text=platform_counts.values
)

fig_platform.update_traces(texttemplate='%{text:.0f}', textposition='outside')
fig_platform.show()
```

По цьому графіку можна побачити, що на Хабрі майже в два рази більше публікацій ніж на Гіктаймі за березень 2015 року

---



```python
day_platform_counts = data_month.groupby(['domain', 'dayofweek']).size().reset_index(name='count')

fig_day_platform = px.line(
    day_platform_counts,
    x='dayofweek',
    y='count',
    color='domain',
    markers=True,
    labels={'dayofweek': 'День місяця', 'count': 'Кількість публікацій'},
    title=f'Кількість публікацій по дням місяця для кожної платформи у {most_published_month}'
)

fig_day_platform.show()
```

По цьому графіку можна побачити, що Хабрі дійсно завжди більше публікацій ніж на Гіктаймсі за березень 2015 року

# Task 3

Коли найкраще опублікувати статтю? Оберіть один або кілька варіантів.

- Найбільше переглядів набирають статті, опубліковані о 12 годині дня

- В опублікованих о 10 ранку постів найбільше коментарів

- Найбільше переглядів набирають статті, опубліковані о 6 годині ранку

- Максимальна кількість коментарів на Гіктаймсі набрала стаття, опублікована о 9
годині вечора

- На Хабрі денні статті коментують частіше, ніж вечірні

```python
import calendar
grouped = df[df['dayofweek'].isin([0, 5])].groupby(['dayofweek', 'hour']).size().reset_index(name="count")
grouped['dayofweek'] = grouped['dayofweek'].replace({i: calendar.day_name[i] for i in range(7)})
fig = px.bar(x=grouped['hour'], y=grouped['count'], color=grouped['dayofweek'],
             labels={'x': 'Година', 'y': 'Кількість публікацій', 'color': 'Day of week'},
             title='Розподіл кількості публікацій по годинах у певні дні тижня',
             barmode='group')

fig.update_xaxes(dtick=1)
fig.show()
```

По цьому графіку можна побачити, що саме більше переглядів публікацій відбувається від 10:00 до 18:00, в цей час користувачі максимально активно переглядають публікації

---



```python
geektimes_night_comments = df[(df['domain'] == 'geektimes.ru') & (df['hour'] == 21)]
max_comments_night = geektimes_night_comments['comments'].max()
print(max_comments_night)
```

Максимальна кількість коментарів Гіктайма о 9 годині вечора

---



```python
# Додавання стовпця 'day_or_night'
df['day_or_night'] = df['hour'].apply(lambda x: 'День' if x < 18 else 'Вечір')

# Підрахунок середньої кількості коментарів для Хабра
habra_comments = df[df['domain'] == 'habrahabr.ru']
avg_comments_day_night = habra_comments.groupby('day_or_night')['comments'].mean().reset_index()

# Побудова графіка для середньої кількості коментарів на Хабрі
fig_habra_comments_day_night = px.bar(
    avg_comments_day_night,
    x='day_or_night',
    y='comments',
    labels={'day_or_night': 'Час доби', 'comments': 'Середня кількість коментарів'},
    title='Середня кількість коментарів на Хабрі в залежності від часу доби'
)

fig_habra_comments_day_night.show()
```

Вніч більша кількість коментарів на Хабрі

# Task 4
Кого з топ-20 авторів найчастіше мінусують?

- @Mordatyj

- @Mithgol

- @alizar

- @ilya42

```python

author_votes_minus = df.groupby('author')['votes_minus'].sum().reset_index()

top_authors = author_votes_minus.nlargest(20, 'votes_minus')

fig_top_authors = px.bar(
    top_authors,
    x='author',
    y='votes_minus',
    labels={'author': 'Автор', 'votes_minus': 'Сумарні негативні оцінки'},
    title='Топ-20 авторів за кількістю негативних оцінок'
)

fig_top_authors.show()

authors_of_interest = ['@Mordatyj', '@Mithgol', '@alizar', '@ilya42']
interest_authors = top_authors[top_authors['author'].isin(authors_of_interest)]

print(interest_authors)
```

# Task 5
Порівняйте суботи і понеділки. Чи правда, що по суботах автори пишуть в основному
вдень, а по понеділках – в основному ввечері?

```python
df['weekday'] = df['published'].dt.day_name()
df['hour'] = df['published'].dt.hour

saturdays = df[df['weekday'] == 'Saturday']
mondays = df[df['weekday'] == 'Monday']

saturdays_hourly = saturdays.groupby('hour').size().reset_index(name='count')
mondays_hourly = mondays.groupby('hour').size().reset_index(name='count')

fig_comparison = px.line(
    pd.concat([saturdays_hourly.assign(weekday='Saturday'), mondays_hourly.assign(weekday='Monday')]),
    x='hour',
    y='count',
    color='weekday',
    markers=True,
    labels={'hour': 'Година доби', 'count': 'Кількість публікацій'},
    title='Публікації по годинах у суботу та понеділок'
)

fig_comparison.show()
```

При порівнянні графіки досить схожі і те, що по суботах автори пишуть в основному вдень, а по понеділках – в основному ввечері, не правда.
В суботу просто менша кількість публікацій ніж в понеділок
